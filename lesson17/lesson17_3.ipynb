```python
#AI建立一個function
#資料來源:scores = np.random.randint(50,101,size=(50, 5))
#欄位: ['國文', '英文', '數學', '自然', '社會','總分','平均','排名']
#索引: ['學生' + str(i) for i in range(1, 51)]
#建立DataFrame
import numpy as np
import pandas as pd 
def create_dataframe():
    scores = np.random.randint(50,101,size=(50, 5))
    columns = ['國文', '英文', '數學', '自然', '社會','總分','平均','排名']
    index = ['學生' + str(i) for i in range(1, 51)]
    df = pd.DataFrame(scores, columns=columns[:-3], index=index)
    df['總分'] = df.sum(axis=1)
    df['平均'] = df.mean(axis=1)
    df['排名'] = df['總分'].rank(ascending=False)
    return df
```

```python
df = create_dataframe()
df
```

```python
#建立1個統計的DataFrame
def create_statistics_dataframe(df):
    cols = ['國文', '英文', '數學', '自然', '社會']
    stats = df[cols].agg(['mean', 'max', 'min', 'median'])
    stats.index = ['平均', '最高', '最低', '中間值']
    return stats
create_statistics_dataframe(df)
```
